```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression  # Importar LinearRegression desde sklearn.linear_model
import matplotlib.pyplot as plt

length = 4 

df = pd.read_csv('data.csv')
datos_subset = pd.read_csv('data.csv')

x = df.metro.values
y = df.precio.values

print(df)
```

```python
# Extracción de las características (metros) y etiquetas (precio) para ambos conjuntos de datos
X_subset = datos_subset['metro']
y_subset = datos_subset['precio']

x = df.metro.values
y = df.precio.values

# Creación y ajuste del modelo de regresión lineal para el primer conjunto de datos (subset)
modelo_lineal_subset = LinearRegression()
modelo_lineal_subset.fit(X_subset.values.reshape(-1, 1), y_subset)

# Obtención de coeficiente e intercepto del modelo para el primer conjunto de datos
coeficiente_subset = modelo_lineal_subset.coef_[0]
intercepto_subset = modelo_lineal_subset.intercept_

print(f"Coeficiente para el subset: {coeficiente_subset}")
print(f"Intercepto para el subset: {intercepto_subset}")

# Realizar una predicción para el primer conjunto de datos
prediccion_subset = modelo_lineal_subset.predict(X_subset.values.reshape(-1, 1))

# Creación y ajuste del modelo de regresión lineal para el segundo conjunto de datos
regr = LinearRegression()
regr.fit(x.reshape(-1, 1), y)

# Obtención de coeficiente e intercepto del modelo para el segundo conjunto de datos
coeficiente_regr = regr.coef_[0]
intercepto_regr = regr.intercept_

print("Coeficiente para el segundo conjunto de datos: %s" % coeficiente_regr)
print("Intercepto para el segundo conjunto de datos: %s" % intercepto_regr)

# Realizar una predicción para el segundo conjunto de datos
X_predict = 35
y_predict = regr.predict([[X_predict]])

print("Predicción para el segundo conjunto de datos: Si x=%s, entonces y=%s" % (X_predict, y_predict[0]))

# Graficar los datos y las líneas de regresión para ambos conjuntos de datos
plt.figure(figsize=(12, 5))

# Primer conjunto de datos
plt.subplot(1, 2, 1)
plt.scatter(X_subset, y_subset, label='Datos Originales (Subset)', color='pink')
plt.plot(X_subset, prediccion_subset, color='purple', linewidth=2, label='Línea de Regresión (Subset)')
plt.xlabel('Metro')
plt.ylabel('Precio')
plt.legend()
plt.title('Regresión Lineal para Subset')

# Segundo conjunto de datos
plt.subplot(1, 2, 2)
plt.scatter(x, y, label='Datos Originales', color='black')
plt.plot(x, regr.predict(x.reshape(-1, 1)), color='blue', linewidth=3, label='Línea de Regresión')
plt.scatter(X_predict, y_predict, color='red', marker='x', s=100, label='Predicción')
plt.xlabel('Metro')
plt.ylabel('Precio')
plt.legend()
plt.title('Regresión Lineal para el Segundo Conjunto de Datos')

plt.tight_layout()
plt.show()
```
